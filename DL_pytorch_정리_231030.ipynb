## ReLU 함수
* ReLU(x) = max(x, 0) 그래프

```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
def relu1(a):
    return np.maximum(0, a)

print(relu1(-3))
```

```python
def relu1(a):
    return np.maximum(0, a)

x1 = np.linspace(start=-5, stop=5, num=26)
print(x1)
x2 = relu1(x1)
print(x2)
px.line(x=x1, y=x2)
```

## sigmoid 함수 그래프

```python
def sigmoid(x):
  return 1/ (1 + np.exp(-x))

array = np.arange(-10, 10, 0.1)
plt.axvline(x=0, color='g', linestyle='--', linewidth=1)
plt.axhline(y=0, color='g', linestyle='--', linewidth=1)
plt.plot(array, sigmoid(array))
plt.show()
```

## tanh 함수 그래프
* tanh(x) 함수는 (ex - e-x) / (ex + e-x)   
= (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))   
= np.sinh(x) / np.cosh(x) 와 동일합니다.

```python
# x = np.linspace(-np.pi, np.pi, 201)
x = np.linspace(-4, 4, 1000)
plt.plot(x, np.tanh(x))
plt.axvline(x=0, color='r', linestyle='--', linewidth=1)
plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.axis('tight')
plt.show()
```
